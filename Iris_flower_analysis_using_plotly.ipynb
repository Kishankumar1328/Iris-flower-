```python
import plotly.express as px
```

```python
df=px.data.iris()
fig=px.scatter(df,x="sepal_width",y="sepal_length",color="species")
fig.show()
```

```python
fig_1=px.scatter(df,x="sepal_width",y="sepal_length",color="species",size="petal_length",hover_data=["petal_width"])
fig_1.show()
```

```python
fig_2=px.scatter(df,x="sepal_width",y="sepal_length",color="species",marginal_y="violin",marginal_x="box",trendline="ols",template="simple_white")
fig_2.show()
```

```python
fig_4=px.parallel_coordinates(df,color="species_id",labels={"species_id":"species","sepal_width":"sepal width","sepal_length":"sepal Length","petal_width":"petal width","petal_length":"petal length"},color_continuous_scale=px.colors.diverging.Tealrose,color_continuous_midpoint=2)
fig_4.show()
```

```python
fig_5=px.scatter_3d(df,x="sepal_length",y="sepal_width",z="petal_width",color="petal_length",size="petal_length",size_max=18,symbol="species",opacity=0.8)
fig_5.show()
```

```python
fig_6=px.scatter_matrix(df,dimensions=["sepal_width","sepal_length","petal_width","petal_length"],color="species")
fig_6.show()
```
